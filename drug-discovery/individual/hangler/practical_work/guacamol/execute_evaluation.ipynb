```python
import pandas as pd
results = pd.read_csv("./linear_regression_suite/results/coati_linear_probing.csv")
results
```

```python
import numpy as np

def highlight_max(data):
    attr = 'background-color: yellow'
    is_max = data == data.max()
    return [attr if v else '' for v in is_max]
```

```python
grouped_df = results.groupby(['Dataset', 'Model']).mean(numeric_only=True).reset_index()
# drop column seed
grouped_df = grouped_df.drop(columns=['Seed'])

# store grouped_df in a csv file
grouped_df.to_csv("coati_linear_probing_grouped.csv", index=False)
```

# Store Combined Results as HTML

```python
# Combine all HTML contents into one HTML file
html_output = '<html><head><style>table {border-collapse: collapse; width: 100%;} th, td {border: 1px solid black; padding: 8px; text-align: left;}</style></head><body>'

for name, group in grouped_df.groupby('Dataset'):
    styled = group.style.apply(highlight_max, subset=['AUROC Score','RMSE','Spearman','avgp','davgp']).render()
    html_output += f'<h1>{name}</h1>{styled}'

html_output += '</body></html>'

# Save the combined HTML to a file
with open('combined_results.html', 'w') as file:
    file.write(html_output)
```
